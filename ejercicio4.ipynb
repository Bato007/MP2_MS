```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def gaussian(x):
  return np.exp(-x**2)
```

```python
def showGrahp(result, x):
  plt.plot(result[1], result[2], '.', markersize=5)
  plt.plot(result[1][result[3]], result[2][result[3]], '.', markersize=5)

  plt.plot(x, gaussian(x))

  plt.xlabel('x')
  plt.ylabel('f(x)')
  plt.grid()
```

```python
x = np.linspace(-5, 5, 100)
plt.plot(x, gaussian(x))
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid()
```

```python
def integralMC(n):
  x = np.random.uniform(-5, 5, n)
  y = np.random.uniform(0, 1, n)
  hits = sum(y <= gaussian(x))
  mascara= y <= gaussian(x)
  return (10) * (hits) / float(n), x, y, mascara
```

```python
result = integralMC(100)
showGrahp(result, x)
```
